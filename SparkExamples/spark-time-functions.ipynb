```python
from pyspark.sql import Row

data = [
    Row(customer_id=i, name=f"Customer_{i}", date_str=f"2025-12-{18+i:02d}", email=f"customer{i}@example.com")
    for i in range(1, 11)
]

df = spark.createDataFrame(data)
display(df)
```

```python
from pyspark.sql.functions import col, lit, sum, count, avg, to_date, contains
df = df.withColumn("date", df.date_str.cast("date"))
df = df.withColumn("date_obj", to_date(col("date_str")))
df.show()
```

```python
# Checking the email to contain customer keyword
df.withColumn("customer_contains", contains(col("email"), lit("customer"))).show()
df.withColumn("customer_contains2", col("email").contains("customer")).show()
```
