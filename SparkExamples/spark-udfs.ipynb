```python

from pyspark.sql import SparkSession
from pyspark.sql.functions import col

spark = SparkSession.builder.appName("UDFExample").getOrCreate()

data = [(1, "Alice"), (2, "Bob"), (3, "Charlie")]
df = spark.createDataFrame(data, ["id", "name"])
df.show()
```

### Normal UDF

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

#Create a python function
def to_upper(name):
    return name.upper()

#Register this function as udf
to_upper_udf = udf(to_upper, StringType())

#Apply the udf to a column
df_normal = df.withColumn('name_uppercased', to_upper_udf(col("name")))
df_normal.display()
```

### Pandas UDF


```python
from pyspark.sql.functions import pandas_udf
from pyspark.sql.types import StringType

#Create a python function and udf in single shot
@pandas_udf(StringType())
def to_upper_case(name):
    return name.str.upper()

df_pandas_udf = df.withColumn("name_uppercased_pd", to_upper_case(col("name")))
df_pandas_udf.display()
```
